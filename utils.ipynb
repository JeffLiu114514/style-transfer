```python

from PIL import Image
import matplotlib.pyplot as plt
```

```python
def image_loader(image_name):
    image = Image.open(image_name)
    return image
image_dir = "./plot_images/"
GATYS_IMAGE = image_loader(image_dir + 'GATYS_300steps.png.png')
AdalN_IMAGE = image_loader(image_dir + 'AdalN.png')
finetune_IMAGE = image_loader(image_dir + 'finetune.png')
scratch_IMAGE = image_loader(image_dir + 'scratch_dancing_stylized_picasso.jpg')

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(16, 16))
ax[0][0].axis('off')
ax[0][0].imshow(GATYS_IMAGE)
ax[0][0].title.set_text('GATYS_300steps')
ax[0][1].axis('off')
ax[0][1].imshow(AdalN_IMAGE)
ax[0][1].title.set_text('AdalN_default')
ax[1][0].axis('off')
ax[1][0].imshow(finetune_IMAGE)
ax[1][0].title.set_text('AdalN_finetune')
ax[1][1].axis('off')
ax[1][1].imshow(scratch_IMAGE)
ax[1][1].title.set_text('AdalN_scratch')
plt.show()
```
